```python
import pandas as pd
import os
```

```python
csv_dir = '../dataset/metadata/metadata.csv'
```

```python
base_dir = '/datasets/xeno_canto/491_dataset/'
```

```python
df = pd.read_csv(csv_dir)
```

```python
paths = df['filename'].tolist()
```

```python
paths = [(base_dir + path) for path in paths]
```

```python
no_audio = []
for path in paths:
    if not os.path.isfile(path):
        no_audio.append(path)
```

```python
no_audio = [name.replace('/datasets/xeno_canto/491_dataset/','') for name in no_audio]
```

```python
for audio in no_audio:
    df = df.drop(df.index[df['filename'] == audio], inplace = True)
```
